```python
import pandas as pd
import json
import sqlite3

# 1. Load Orders
orders_df = pd.read_csv('orders.csv')

# 2. Load Users (JSON)
with open('users.json', 'r') as f:
    users_data = json.load(f)
users_df = pd.DataFrame(users_data)   

# 3. Load Restaurants (SQL - converting to DataFrame)
conn = sqlite3.connect(':memory:')
restaurants_df = pd.read_csv('restaurants.csv')
```

```python
merged_df = pd.merge(orders_df, users_df, on='user_id', how='left')

final_df = pd.merge(merged_df, restaurants_df, on='restaurant_id', how='left')

final_df.to_csv('final_food_delivery_dataset.csv', index=False)
print("Data successfully merged and saved!")
```

```python
gold_aov = final_df[final_df['membership'] == 'Gold']['total_amount'].mean()
print(f"Average Order Value (Gold): {gold_aov:.2f}")

high_rating_count = len(final_df[final_df['rating'] >= 4.5])
print(f"High Rating Orders: {high_rating_count}")

gold_city_revenue = final_df[final_df['membership'] == 'Gold'].groupby('city')['total_amount'].sum()
top_city = gold_city_revenue.idxmax()
gold_orders_in_top_city = len(final_df[(final_df['membership'] == 'Gold') & (final_df['city'] == top_city)])
print(f"Orders in {top_city} (Top Gold City): {gold_orders_in_top_city}")
```
